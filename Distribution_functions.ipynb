```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import beta
import warnings
warnings.filterwarnings("ignore")
alpha,Beta,n,bin =2,2,6000,40
N = np.random.beta(alpha,Beta,n)
plt.figure(1)
nn,bins,x = plt.hist(N,bins=bin,density=True,color = 'white',edgecolor = 'black', label=r'$N$ = '+str(N.size) + r',$\alpha$ ='+str(alpha)+ r',$\beta$ ='+str(Beta) + ',\n$\mu$ ='+str(round(np.mean(N),3))+ ',$\sigma^2$ ='+str(round(np.var(N),2)))
plt.plot(bins,(bins**(alpha-1))*((1-bins)**(Beta-1))/beta(alpha,Beta),label=r'$P(x) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}$'+';\n'+r'$\mu =\frac{\alpha}{\alpha+\beta} = $'+str(round(alpha/(alpha + Beta),2))+ r',$\sigma^2 =\frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}= $'+str(round((alpha*Beta)/((alpha + Beta)**2*(alpha+Beta+1)),2)))
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.xlim([min(bins), max(bins)]); plt.xlabel('x')
plt.ylabel(r'$P_{beta}(x)$')
plt.show()
def moment(X,n):
    mu = np.mean((X-np.mean(X))**n)
    return mu
for i in range(4):
    print(f'mu_{i} about the mean = {round(moment(N,i),3)}')
print(f'cumulant k_1 = {round(moment(N,1),3)}')
print(f'cumulant k_2 = {round(np.var(N),3)}')
print(f'cumulant k_3 = {round(moment(N,3) - 3*moment(N,1)*moment(N,2) + 2*moment(N,1)**3,3)}')
print(f'cumulant k_4 = {round(moment(N,4) - 4*moment(N,1)*moment(N,3) - 3*moment(N,2)**2 + 12*moment(N,2)*moment(N,1)**2 - 6*moment(N,1)**4,3)}')
```

```python
import numpy as np
from scipy.misc import derivative

# Define the MGF function
def mgf(t):
    return np.exp(t) / (1 - t)  # Example MGF

# Compute the nth derivative of the MGF at t=0
def nth_derivative(mgf_func, t=0, n=1):
    return derivative(mgf_func, t, dx=1e-6, n=n, order=n*2+1)

# Compute moments
first_moment = nth_derivative(mgf, n=1)
second_moment = nth_derivative(mgf, n=2)
third_moment = nth_derivative(mgf, n=3)

print(f"First Moment (Mean): {first_moment}")
print(f"Second Moment: {second_moment}")
print(f"Third Moment: {third_moment}")
```
