```python
import numpy as np
import matplotlib.pyplot as plt
def rand_dir():
    theta = np.random.choice([0,np.pi/2,np.pi,1.5*np.pi])
    r = np.array([np.cos(theta),np.sin(theta)])
    return r
def rand_dir_array(n,flg = False):
    res = []
    f = np.array([0,0])
    for i in range(n):
        a = rand_dir()
        f = f+a
        if not flg:
            res.append(f)
        else:
            res.append(f[0]**2 + f[1]**2)
    return np.array(res)
res = rand_dir_array(1000)
plt.plot(res[:,0],res[:,1])
plt.grid()
plt.show()
config,steps = 1000,1000
walks = np.array([rand_dir_array(steps,True) for i in range(config)])
rms = np.log(np.sqrt(np.mean(walks,axis = 0)))
t = np.log(np.arange(1,config+1))
plt.figure(2)
plt.plot(t,rms,label = 'calculated')
plt.plot(t,0.5*t,label = 'theoretical')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
N,T = 1000,1000
t = np.arange(1,T+1)
walks = 2* np.random.randint(2,size=(N,T))-1
position = np.cumsum(walks,axis = 1)
pos_sq = position**2
mean_pos_sq = np.mean(pos_sq,axis=0)
rms = np.sqrt(mean_pos_sq)
t = np.log(t)
rms = np.log(rms)
coef = poly.polyval(t,rms,1)
rmsfit = poly.polyval(t,coef)
plt.plot(t,0.5*t,t,rms)
plt.legend(['calculated value','theoretical value'])
plt.grid()
plt.title('Random walk')
plt.xlabel(r'$log_{e}(T)$')
plt.ylabel(r'$log_{e}(x_{rms})$')
plt.show()
```
